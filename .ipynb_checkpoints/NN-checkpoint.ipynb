```python
from sklearn.model_selection import train_test_split
import numpy as np

X = np.load('/home/jbmai_sai/Desktop/climate/train_x.npy') 
   
y =  np.load('/home/jbmai_sai/Desktop/climate/train_y.npy')
    
#X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
n_cols = X.shape
n_cols
```

```python
from keras.models import Sequential
from keras.layers import Dense
#create model
model = Sequential()

#get number of columns in training data
n_cols = X.shape[1]

#add model layers
model.add(Dense(250, activation='relu', input_shape=(n_cols,)))
model.add(Dense(125, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='relu'))

model.compile(optimizer='adam', loss='mean_absolute_error')
```

```python
from keras.callbacks import EarlyStopping
#set early stopping monitor so the model stops training when it won't improve anymore
early_stopping_monitor = EarlyStopping(patience=3)
#train model
model.fit(X,y, validation_split=0.15, epochs=30, callbacks=[early_stopping_monitor])
```

```python
Xt = np.load('/home/jbmai_sai/Desktop/climate/test_x.npy') 

#X[0:1]



y = model.predict(Xt)
```
