```python
import pandas as pd
df = pd.read_csv('/home/jbmai_sai/Desktop/climate/train_dataset.csv')

print(df.isnull().sum())

print(df.isna().sum())
```

```python


df = pd.concat([df,pd.get_dummies(df['weather_type'], prefix='weather_type')],axis=1)

for i in df.index:
    day = df['is_holiday'].loc[i]
    if(day != 'None'):
        df['is_holiday'].loc[i] = 'holiday'
        
df = pd.concat([df,pd.get_dummies(df['is_holiday'], prefix='is_holiday')],axis=1)

df = pd.concat([df,pd.get_dummies(df['weather_description'], prefix='weather_description')],axis=1)

new = df["date_time"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
df["date"]= new[0] 
  
# making separate last name column from new data frame 
df["time"]= new[1] 

new = df["date"].str.split("-", n = 2, expand = True) 
  
# making separate first name column from new data frame 
df["year"]= new[0] 
  
# making separate last name column from new data frame 
df["month"]= new[1] 
df["day"] = new[2]
  
# Dropping old Name columns 
new = df["time"].str.split(":", n = 1, expand = True) 
  
# making separate first name column from new data frame 
df["hour"]= new[0] 
  
```

```python
df.drop(columns = ['weather_description','date_time','date','time','is_holiday','weather_type'], inplace = True)
d = df
df_y = df['traffic_volume']

df.drop(columns = ['traffic_volume'],inplace = True)
d = df
```

```python
df["year"] = pd.to_numeric(df["year"])
df["month"] = pd.to_numeric(df["month"])
df["day"] = pd.to_numeric(df["day"])
df["hour"] = pd.to_numeric(df["hour"])
```
